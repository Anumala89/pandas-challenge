```python
# Dependencies and Setup
import pandas as pd
import os 

# File to Load (Remember to Change These)
school_data_to_load = os.path.join("Resources/schools_complete.csv")
student_data_to_load =os.path.join("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
school_data_complete.head()
```

## District Summary

```python
#Calculate the total number of schools
total_school = school_data_complete["school_name"].nunique()
#total_school
#Calculate the total number of students
total_student = school_data_complete["student_name"].count()
#total_student
#Calculate the total budget
total_budget = (school_data_complete["budget"].unique()).sum()
#total_budget
```

```python
#Calculate the average math score 
avg_math_score = round((school_data_complete["math_score"]).mean(),6)
#avg_math_score
#Calculate the average reading score
avg_reading_score = round((school_data_complete["reading_score"]).mean(),6)
#avg_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math =  school_data_complete.loc[school_data_complete["math_score"] >= 70, ["student_name"]]
passing_math_percent = round((passing_math["student_name"].count()/total_student)*100, 6)
#passing_math_percent

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70, ["student_name"]]
passing_read_percent = round((passing_read["student_name"].count()/total_student)*100, 6)
#passing_read_percent

#Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_overall = passing_math + passing_read
passing_overall_percent = round((passing_overall["student_name"].count() / total_student)*100, 6)
#passing_overall_percent
```

```python
#create a dataframe form the results 
summary_df = pd.DataFrame({"Total Schools" :[total_school], "Total Students" :[total_student],
                           "Total Budget" :[total_budget], "Average Math Score" :[avg_math_score],
                          "Average Reading Score" :[avg_reading_score], "% Passing Math" :[passing_math_percent],
                          "% Passing Reading" :[passing_read_percent], "% Overall Passing" :[passing_overall_percent]})

#format the numbers to display
summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)
summary_df
```

## School Summary

```python
#get school name and type for each school in order to create school summary
school_data = school_data_complete.groupby("school_name")
#school_data.head()

school_type = school_data["type"].unique()
#school_type
```

```python
#get the total student and the budget for each school
total_students2 = school_data["size"].unique()
#total_students2

total_budget2 = school_data["budget"].unique()
#total_budget2
```

```python
#get the budget per student and average scores for each school
per_student_budget = total_budget2 / total_students2
#per_student_budget

avg_math_score2 = school_data["math_score"].mean()
#avg_math_score2

avg_reading_score2 = school_data["reading_score"].mean()
#vg_reading_score2
```

```python
#get the passing percentages for each school
passing_math2 = school_data["math_score"].apply(lambda x: ((x >= 70)/len(x)).sum())
passing_math2_percent2 = round(passing_math2 * 100 , 6)
#passing_math2_percent2

passing_read2 = school_data["reading_score"].apply(lambda x: ((x >= 70)/len(x)).sum())
passing_read2_percent2 = round(passing_read2 * 100 , 6)
#passing_read2_percent2

overall = school_data_complete.loc[(school_data_complete["reading_score"].values >= 70) & (school_data_complete["math_score"].values >= 70)]
overall_passing2= overall.groupby("school_name")
overall_passing2_percent2 =(overall_passing2["school_name"].count()/ total_students2)*100
#overall_passing2_percent2
```

```python
#create a school summary dataframe form the results 
df2 = pd.DataFrame({"School Type": school_type, "Total Students": total_students2, "Total School Budget": total_budget2,
                    "Per Student Budget": per_student_budget, "Average Math Score" : avg_math_score2, 
                    "Average Reading Score": avg_reading_score2, "% Passing Math": passing_math2_percent2,
                    "% Passing Reading": passing_read2_percent2, "% Overall Passing" : overall_passing2_percent2})

#format the numbers to display
df2["School Type"] = df2["School Type"].str[0]
df2["Total Students"] = df2["Total Students"].str[0]
df2["Total Students"] = df2["Total Students"].map("${:,.2f}".format)
df2["Total School Budget"] = df2["Total School Budget"].str[0]
df2["Total School Budget"] = df2["Total School Budget"].map("${:,.2f}".format)
df2["Per Student Budget"] = df2["Per Student Budget"].str[0]
df2["% Overall Passing"] = df2["% Overall Passing"].str[0]
df2
```

## Top Performing School ( By % Overall Passing

```python
#get the top five performing schools
top_five=df2.sort_values(by = "% Overall Passing", ascending = False)
top_five.head()
```

## Bottom Performing Schools (By %Overall Passing)

```python
#get the worst five performing school
worst_five = df2.sort_values(by ="% Overall Passing", ascending = True)
worst_five.head()
```

## Math Scores by Grade

```python
#get the unique values in the grade column for the math score by grade
grade = school_data_complete["grade"].unique()
#grade

#get the scores by the grade
ninth = school_data_complete.loc[school_data_complete["grade"] == "9th",["school_name","math_score"]]
math9 = ninth.groupby("school_name")["math_score"].mean()
tenth = school_data_complete.loc[school_data_complete["grade"] == "10th",["school_name","math_score"]]
math10= tenth.groupby("school_name")["math_score"].mean()
eleventh = school_data_complete.loc[school_data_complete["grade"] == "11th",["school_name","math_score"]]
math11 = eleventh.groupby("school_name")["math_score"].mean()
twelfth = school_data_complete.loc[school_data_complete["grade"] == "12th",["school_name","math_score"]]
math12 = twelfth.groupby("school_name")["math_score"].mean()

#create a dataframe from the results
df3 = pd.DataFrame({"9th": math9, "10th": math10, "11th": math11, "12th": math12})
df3
```

## Reading Scores by Grade

```python
#get the scores by the grade for the reading score by grade
Ninth = school_data_complete.loc[school_data_complete["grade"] == "9th",["school_name","reading_score"]]
Math9 = Ninth.groupby("school_name")["reading_score"].mean()
Tenth = school_data_complete.loc[school_data_complete["grade"] == "10th",["school_name","reading_score"]]
Math10= Tenth.groupby("school_name")["reading_score"].mean()
Eleventh = school_data_complete.loc[school_data_complete["grade"] == "11th",["school_name","reading_score"]]
Math11 = Eleventh.groupby("school_name")["reading_score"].mean()
Twelfth = school_data_complete.loc[school_data_complete["grade"] == "12th",["school_name","reading_score"]]
Math12 = Twelfth.groupby("school_name")["reading_score"].mean()

#create a dataframe from the results
df4 = pd.DataFrame({"9th": Math9, "10th": Math10, "11th": Math11, "12th": Math12})
df4
```

## Scores by School Spending

```python
#get the related columns
spending = school_data_complete.loc[:,("budget", "math_score", "reading_score")]
spending["Per_Student_Budget"] = school_data_complete["budget"].values/ school_data_complete["size"]

#create the range 
bins = [0, 585, 630, 645, 675]
range_names =["<$584", "$585-629", "$630-644", "$645-675"]

#place the range column in the dataframe
spending["Spending Ranges (Per Student)"] = pd.cut(spending["Per_Student_Budget"],bins, labels = range_names, right = False)
spending_data = spending.groupby("Spending Ranges (Per Student)")
```

```python
#get the average scores
avg_math3 = round(spending_data["math_score"].mean(), 2)
avg_reading3 = round(spending_data["reading_score"].mean(), 2)

#get the percentages of passing
passing_math3 = spending_data["math_score"].apply(lambda x: ((x >= 70)/len(x)).sum())
passing_math3_percent3 = round(passing_math3 * 100 , 2)
#passing_math3_percent3

passing_read3 = spending_data["reading_score"].apply(lambda x: ((x >= 70)/len(x)).sum())
passing_read3_percent3 = round(passing_read3 * 100 , 2)
#passing_read3_percent3

overall3 = spending.loc[(spending["reading_score"].values >= 70) & (spending["math_score"].values >= 70)]
overall_passing3= overall3.groupby("Spending Ranges (Per Student)")
#overall_passing3.head()
overall_passing3_percent3 = round((overall_passing3["Per_Student_Budget"].count()/spending_data["budget"].count())*100, 2)
#overall_passing3_percent3


#create a dataframe from the results
df5= pd.DataFrame({"Average Math Score": avg_math3, "Average Reading Score": avg_reading3, 
                    "% Passing Math": passing_math3_percent3, "% Passing Reading": passing_read3_percent3,
                   "% Overall Passing": overall_passing3_percent3})
df5
```

## Scores by School Size

```python
#get the related columns
size = school_data_complete.loc[:,("size", "math_score", "reading_score")]

#create the range
bins = [0, 1000, 2000, 5000]
size_range =["Small(<1000)", "Medium(1000-2000)", "Large(2000-5000)"]

#place the range column in the dataframe
size["School Size"] = pd.cut(size["size"],bins, labels = size_range , right = False )
size_data = size.groupby("School Size")
```

```python
#get the average score
avg_math4 = round(size_data["math_score"].mean(), 2)
avg_reading4 = round(size_data["reading_score"].mean(), 2)

#get the percentages of passing
passing_math4 = size_data["math_score"].apply(lambda x: ((x >= 70)/len(x)).sum())
passing_math4_percent4 = round(passing_math4 * 100 , 2)
#passing_math4_percent4

passing_read4 = size_data["reading_score"].apply(lambda x: ((x >= 70)/len(x)).sum())
passing_read4_percent4 = round(passing_read4 * 100 , 2)
#passing_read4_percent4

overall4 = size.loc[(size["reading_score"].values >= 70) & (size["math_score"].values >= 70)]
overall_passing4= overall4.groupby("School Size")
#overall_passing4.head()
overall_passing4_percent4 = round((overall_passing4["School Size"].count()/size_data["size"].count())*100, 2)
overall_passing4_percent4

#create a dataframe from the results
df6= pd.DataFrame({"Average Math Score": avg_math4, "Average Reading Score": avg_reading4, 
                    "% Passing Math": passing_math4_percent4, "% Passing Reading": passing_read4_percent4,
                   "% Overall Passing": overall_passing4_percent4})
df6
```
